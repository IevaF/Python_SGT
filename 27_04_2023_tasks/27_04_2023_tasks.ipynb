```python
import pandas as pd
import mysql.connector

db = mysql.connector.connect(user='root',
                              password='',
                              host='localhost',
                              database='countries_data')
```

```python
countries_df = pd.read_excel('C:\\Ieva\\She goes Tech\\Task_folder\\28_04_2023_tasks\Countries.xlsx')
countries_df
```

```python
cursor = db.cursor()

for i, row in countries_df.iterrows():
    cursor.execute("SELECT Country_ID FROM countries WHERE Country_ID = %s", (row['Country_ID'],))
    result = cursor.fetchone()
    
    if result:
        cursor.execute("UPDATE countries SET Country = %s, Area = %s, Population = %s, Capital = %s WHERE Country_ID = %s",
                       (row['Country'], row['Area'], row['Population'], row['Capital'], row['Country_ID']))
    
    else:
        cursor.execute("INSERT INTO countries (`Country_ID`, `Country`, `Area`, `Population`, `Capital`) VALUES (%s, %s, %s, %s, %s)",
                       (row['Country_ID'], row['Country'], row['Area'], row['Population'], row['Capital']))
        
   
    db.commit()
```

```python
for i, row in cities_df.iterrows():
    cursor.execute("SELECT Country_ID FROM countries WHERE Country_ID = %s", (row['Country_City_ID'],))
    result = cursor.fetchone()
    
    if result:
        cursor.execute("SELECT City_ID FROM cities WHERE City_ID = %s", (row['City_ID'],))
        result = cursor.fetchone()
        
        if result:
            cursor.execute("UPDATE cities SET City = %s, Population = %s WHERE City_ID = %s", (row['City'], row['Population'], row['City_ID']))
        else:
            cursor.execute("INSERT INTO cities (City_ID, City, Population, Country_City_ID) VALUES (%s, %s, %s, %s)", (row['City_ID'], row['City'], row['Population'], row['Country_City_ID']))
        
db.commit()
```

```python
import pandas as pd
import mysql.connector

db = mysql.connector.connect(user='root',
                              password='',
                              host='localhost',
                              database='countries_data')

cursor = db.cursor()

cursor.execute("SELECT * FROM countries")
result = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
countries_df = pd.DataFrame(result, columns=columns)
countries_df.to_excel(r'C:\Ieva\She goes Tech\Task_folder\28_04_2023_tasks\countries_export.xlsx', index=False)

cursor.execute("SELECT * FROM cities")
result = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
cities_df = pd.DataFrame(result, columns=columns)
cities_df.to_excel(r'C:\Ieva\She goes Tech\Task_folder\28_04_2023_tasks\cities_export.xlsx', index=False)

db.close()
```
