```python
#4) Use Pandas to group the data by a categorical variable and calculate summary statistics for each group.
```

```python
# Values are from year 2022
# Groups:
A = "industry, construction, transportation"
B = "Services"
C = "Education, art, culture, others"

group_A_C = pandas.DataFrame({'Category': ['A','A','A','B','B','A','B','B','B','B','B','B','B','B','B','C','C','C','C'],
                   'Value': [10.34, 11.68, 10.79, 14.85, 11.2, 10.53, 9.77, 10.57, 7.20, 18.71, 20.64, 8.96, 13.82, 9.80, 14.16, 10.19, 13.89, 9.95, 8.67]})
                    
grouped_data = group_A_C.groupby('Category').agg({'Value': ['min', 'max', 'median','mean', 'std']})
print(grouped_data)
```

```python
#7) Create a box plot to visualize the distribution of a variable across different categories.
```

```python
years = ['2019','2020','2021','2022']
gads_2019 = data['2019']
gads_2020 = data['2020']
gads_2021 = data['2021']
gads_2022 = data['2022']

data = [gads_2019, gads_2020, gads_2021, gads_2022]

pl.boxplot(data, labels=years)

pl.title("Box Plot - average labor costs per hour from 2019-2022")
pl.xlabel("Year")
pl.ylabel("Av_labor costs per hour")

pl.show()
```
